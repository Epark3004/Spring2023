```python
#@markdown ðŸ”„ Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ctts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "ca",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")
```

# 'r' sound in different positions

## A. In stressed syllables

```python
positions = "1_word-initial" #@param = ["1_word-initial", "2_word-medial", "3_word-final"]
accents = "BrE" #@param = ["AmE","BrE"]

pos = positions.split("_")
num = pos[0]
n=int(num)-1

wordlist = ["earth, earn, irk, herb, early, urgent, urban",
            "turn, burn, learn, heard, verb, word, curve, third, Thursday, circus, bird, firm, hurt, term",
            "sir, her, fur, prefer, defer, occur"]
wd = wordlist[n]
intro2 = pos[1]
mytxt = "Let me say some English words appearing in %s. Listen how the R sound is pronounced. "%intro2 + wd

if accents == "AmE":
  etts(mytxt)
elif accents == "BrE":
  btts(mytxt)

Audio("E-audio.mp3", autoplay = True)
```

## B. In unstressed syllables

```python
positions = "1_word-initial" #@param = ["1_word-initial", "2_word-medial", "3_word-final"]
accents = "BrE" #@param = ["AmE","BrE"]

pos = positions.split("_")
num = pos[0]
n=int(num)-1

wordlist = ["No English word is found containing this syllable in this position",
            "Words in medial position: turn, burn, learn, heard, verb, word, curve, third, Thursday, circus, bird, firm, hurt, term",
            "Words in final position: sir, her, fur, prefer, defer, occur"]
wd = wordlist[n]
intro2 = pos[1]
mytxt = wd

if accents == "AmE":
  etts(mytxt)
elif accents == "BrE":
  btts(mytxt)

Audio("E-audio.mp3", autoplay = True)
```

# How to pronounce '-rl' together?

+ rl:  "girl, curl, world, pearl, early"
+ r:   "sir, fur, word, per, err",
+ l:   "gull, cull, wall, paul, all"

```python
#@markdown girl, curl, world, pearl, early

words = "1_words with rl" #@param = ["1_words with rl", "2_ words with r", "3_words with l"]
accents = "American" #@param = ["American", "British"]
wd = words.split("_")
at = accents.split(" ")

num = int(wd[0])
n = num -1

wordlist = ["girl, curl, world, pearl, early", 
            "sir, fur, word, per, err",
            "gull, cull, wall, paul, all"]

txt = "Here's the word sample: " + wordlist[n]

if accents == "American":
  etts(txt)
elif accents == "British":
  btts(txt)

Audio("E-audio.mp3", autoplay = True)
```
