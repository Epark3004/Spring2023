```python
#@markdown üí† Import libraries {Image, Audio, YouTubeVideo}
from IPython.display import Image, Audio, YouTubeVideo
```

```python
#@markdown üîÑ Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = True)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = True)
  
  gtts_object.save("B-audio.mp3")
  return Audio("B-audio.mp3")

def ctts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "ca",
                    slow = True)
  
  gtts_object.save("C-audio.mp3")
  return Audio("C-audio.mp3")
```

# [1] Korean vs. English vowels: [  å ], [ …î ], and [ o ä ]

```python
#@markdown Korean accent:
Spelling = "2" #@param = [1, 2, 3, 4, 5]

pn = int(Spelling) -1

wordlist = ["but, bought, and boat",
         "cut, caught, and coat",
         "hull, hall, and hole",
         "colour, caller, and collar",
         "mull, mall, and mole"]

wlist = wordlist[pn]

txt = "Ï†úÍ∞Ä ÏÑúÎ°ú Îã§Î•∏ ÏòÅÏñ¥ Îã®Ïñ¥ ÏÑ∏ Í∞úÎ•º ÎßêÌï¥Î≥ºÍ≤åÏöî." + " " + wlist

ktts(txt)
Audio("K-audio.mp3", autoplay = True)
```

```python
#@markdown English accent:
Spelling = "1" #@param = [1, 2, 3, 4, 5]

pn = int(Spelling) -1

wordlist = ["but, bought, and boat",
         "cut, caught, and coat",
         "hull, hall, and hole",
         "colour: caller, and collar",
         "mull, mall, and mole"]

wlist = wordlist[pn]
txt = "Let me say three different English words. Listen carefully. " + " " + wlist
etts(txt)
Audio("E-audio.mp3", autoplay = True)
```

# A. Warming up: [  å ]

```python
#

Positions = "3-At the end of words" #@param = ["1-At the beginning of words", "2-In the middle of words", "3-At the end of words"]

ps = Positions.split('-')
np = ps[0]
np = int(np) -1

wordlist = ["up, of, us, uncle, under, upper, onion, other, ugly",
            "lucky, hug, much, rough, does, come, trouble, done, nut, must, sun, son, some",
            "There is no word ending with this vowel in English."]

wlist = wordlist[np]
intro = ps[1] + ":" 
mytxt = intro + wlist + "." + "This is the end."
etts(mytxt)

Audio("E-audio.mp3", autoplay = True)

```

# B. Warming up: [ …î ]

```python
#

Positions = "3-At the end of words" #@param = ["1-At the beginning of words", "2-In the middle of words", "3-At the end of words"]

ps = Positions.split('-')
np = ps[0]
np = int(np) -1

wordlist = ["off, all, also, August, audience, ought, often, always",
            "boss, long, across, song, bought, caught, broad, taught, wrong, dog, toss, lawn, auto, auction, mall, call, fall, salt, stall.",
            "law, raw, jaw, straw, draw, flaw."]

wlist = wordlist[np]
intro = ps[1] + ":" 
mytxt = intro + wlist + "." + "This is the end."
etts(mytxt)

Audio("E-audio.mp3", autoplay = True)

```
