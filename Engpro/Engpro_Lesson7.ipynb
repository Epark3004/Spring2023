```python
#@markdown üîÑ Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = True)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = True)
  
  gtts_object.save("B-audio.mp3")
  return Audio("B-audio.mp3")

def ctts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "ca",
                    slow = True)
  
  gtts_object.save("C-audio.mp3")
  return Audio("C-audio.mp3")
```

```python
#@markdown English accent:
Spelling = "1" #@param = [1, 2, 3, 4, 5]

pn = int(Spelling) -1

wordlist = ["game, may, say, hey, name",
         "night, hi, bye, sigh, guy",
         "how, now, hour, sour, loud",
         "don't, no, so, cold, hold",
         "boy, oil, noise, choice, soil"]

wlist = wordlist[pn]
txt = "Let me say five different English words. Listen carefully. " + " " + wlist
etts(txt)
Audio("E-audio.mp3", autoplay = True)
```

# [2] Diphthong vowels in sentences

```python
#

Positions = "1" #@param = [1, 2, 3, 4, 5]

np = int(Positions) -1

wordlist = ["Oh, really?",
            "So, we are finished!",
            "Please don't tell him.",
            "It's very cold in here.",
            "You don't have to."]

wlist = wordlist[np]
intro = "Number " + Positions + "!" 
mytxt = intro + wlist
etts(mytxt)

Audio("E-audio.mp3", autoplay = True)

```
