### gTTS setting

```python
#@markdown 🔄 Making functions: etts('text'), ktts('text')
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")
```

## [2] Accuracy, Intelligibility, and comprehensibility

+ **Accuracy** of speech refers to the **correctness of pronunciation, grammar, and vocabulary** in a speaker's language use. It measures adherence to standard norms and is important for effective communication and being understood.
+ **Intelligibility** of speech refers to the **clarity of a speaker's words and their ability to be understood** by listeners. It is concerned with effective communication, not just language accuracy.
+ **Comprehensibility** is the **ease of understanding** a speaker's message, considering context, background, and language use. It evaluates the overall communication process, including understanding of the speaker's intent, not just speech clarity. It is broader than intelligibility, which only focuses on pronunciation and articulation.
