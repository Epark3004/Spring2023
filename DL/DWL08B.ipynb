```python
#@markdown Setup: import libraries
!pip install gtts
from gtts import gTTS
from IPython.display import Audio

def tts(txt):
    tts = gTTS(txt, lang='en')
    tts.save("sample.mp3")
    return Audio("sample.mp3", autoplay=True)
```
