# [2] Spoken response from chatGPT using gTTS

```python
#@markdown üö© Install and import libraries {gtts}{Audio}
%%capture
!pip install gtts
from gtts import gTTS
from IPython.display import Audio
```

```python
#@markdown üö© Define tts(text):
def tts(text):
  tts = gTTS(txt, lang='en')
  tts.save("myaudio.mp3")
  return Audio("/content/myaudio.mp3")
```

## ChatGPT Response as audioplay

```python
#@markdown Instruction (audio)
name = input("Name: ")
intro = """
Let me explain how you can use me in the application below.
When you run the code below, you'll see a textbox.
You can ask any question in English, and I'll tell you the answer.
If you are ready to start, run the code below.
"""
txt = "Hello %s!"%name + intro
tts(txt)

Audio("myaudio.mp3", autoplay = True)
```

```python
#@markdown User vs. ChatGPT
print("üòÉ Okay, %s. Type your question next to your name in the box"%name)
print("="*80)

question = input("%s: "%name)
completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "user","content": question}
     ]
)
print("*"*80)
chat_response = completion.choices[0].message.content
print(f'ChatGPT: {chat_response}')
print("*"*80)
voice = input("üòÉ Do you want me to say the answer? ('y' or 'n'): ")
print("*"*80)
print("üòÉ Note: it may take a few seconds if the answer is long. Be patient!")
print("*"*80)
if voice == "y":
  txt = chat_response
  tts(txt)
elif voice == "n":
  txt = "Okay then. I'd say nothing."
  tts(txt)
Audio("myaudio.mp3", autoplay = True)
```

# [3] System role

+ This can change the response style of chatGPT: e.g., friendly assistant, recruiter who asks tough questions, etc.

```python
# Define "content" style (response style)
messages = [
    {"role": "system", "content": "You're a kind helpful English as Second Language teacher."}
]
```

```python
#@markdown Q & A (text):
content = input("User: ")
messages.append({"role": "user", "content": content})

completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = messages
)

chat_response = completion.choices[0].message.content
print(f'ChatGPT: {chat_response}')
```

```python
#@markdown Q & A (spoken)
content = input("User: ")
messages.append({"role": "user", "content": content})

completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = messages
)

chat_response = completion.choices[0].message.content
# print(f'ChatGPT: {chat_response}')

txt = chat_response
tts(txt)
Audio("myaudio.mp3")
```
