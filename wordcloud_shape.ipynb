```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Get user input
user_text = input("Enter your text: ")

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(user_text)

# Display the generated image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

# Save the wordcloud image
plt.savefig("wordcloud.png")

# Show the image
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Get user input
user_text = input("Enter your text: ")

# List of keywords to emphasize
keywords = ["accessibility", "personalization", "collaboration", "data-driven", "life-long learning", "multimodality"]

# Repeat the keywords multiple times to make them stand out
for keyword in keywords:
    user_text += " " + (keyword + " ") * 10

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(user_text)

# Display the generated image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

# Save the wordcloud image
plt.savefig("wordcloud.png")

# Show the image
plt.show()
```

[GNU image](https://github.com/MK316/workshops/raw/main/20230512_GNU/images/GNU.png)

```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, STOPWORDS
from PIL import Image, ImageOps

# Get user input
user_text = input("Enter your text: ")

# Load the "GNU" image as a mask
mask_image = Image.open("GNU.png").convert("L")
mask_image = ImageOps.invert(mask_image)  # Invert the mask image
mask = np.array(mask_image)

# Create a WordCloud object with the custom mask and black background
wordcloud = WordCloud(width=800, height=800, background_color='black', min_font_size=10, mask=mask,
                      contour_width=3, contour_color='turquoise').generate(user_text)

# Display the generated image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

# Save the wordcloud image
plt.savefig("wordcloud_gnu.png")

# Show the image
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, STOPWORDS
from PIL import Image, ImageOps

# Get user input
user_text = input("Enter your text: ")

# Replace "-" linked words with underscores "_"
user_text = user_text.replace("-", "_")

# Load the "GNU" image as a mask
mask_image = Image.open("multimodal.png").convert("L")
mask_image = ImageOps.invert(mask_image)  # Invert the mask image
mask = np.array(mask_image)

# Create a WordCloud object with the custom mask and black background
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10, mask=mask,
                      contour_width=3, contour_color='turquoise').generate(user_text)

# Display the generated image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

# Save the wordcloud image
plt.savefig("wordcloud_gnu.png")

# Show the image
plt.show()
```
