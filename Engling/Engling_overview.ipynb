# Chapter 3. Overivew audio

```python
#@markdown Settingup:
%%capture
!pip install gtts
from gtts import gTTS
from IPython.display import Audio
import pandas as pd

dialect = "us"
def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = dialect,
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")
def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")
```

```python
#@markdown Data to read
url = "https://raw.githubusercontent.com/MK316/Spring2023/main/Engling/Engling_chapteroverview.csv"

data = pd.read_csv(url)
data = data.iloc[:5]
data
```

```python
#@markdown Select passages:
parts = "0" #@param = [0,1,2,3,4]

txt = data['Contents'][0]
etts(txt)
Audio("E-audio.mp3", autoplay=True)
```
