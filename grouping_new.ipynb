## [1] Import packages, read a file 

Note: Column name as "Names"

```python
#@markdown Import packages
import pandas as pd
import numpy as np
import random
```

+ sample: https://raw.githubusercontent.com/MK316/Spring2023/main/data/att_sample.csv
+ DL: https://raw.githubusercontent.com/MK316/Spring2023/main/DL/files/DL_attfinal.csv

```python
#@markdown Read attendance list, define grouping(l, n)
url = input()
df = pd.read_csv(url)
attlist = df.Names.values.tolist()

def grouping(l, n):
    for i in range(0, len(l), n):
        yield l[i:i + n]

df.tail()
```

## [2] Random grouping

```python
#@markdown 🌀 Random grouping of size "n"
random.shuffle(attlist)

gsize = input("Size of each group: ")
gs = int(gsize)

mylist = list(grouping(attlist, gs))
ngroups = len(list(grouping(attlist, gs)))

print(mylist)
print("="*50)
print("Number of groups: %d in total."%ngroups)

gn = ngroups
gname = []
df1 = pd.DataFrame()
for i in range(0,gn):
  nam = "G" + str(i+1)
  gname.append(nam)

df1['Group'] = gname
df1['Members'] = mylist

df1
```

```python
#@markdown Group members in csv: column modification (df2)
n = input('Number of groups: ')
n = int(n)
df2 = pd.DataFrame()
nlist1 = []
nlist2 = []
nlist3 = []
nlist4 = []
group = []

for i in range(0,n):
  gn = str(i+1)
  gr = 'G'+ gn
  group.append(gr)


for i in range(0,n):
    list = df1['Members'][i]
    if len(list) == 4:
        n1 = df1['Members'][i][0]
        n2 = df1['Members'][i][1]
        n3 = df1['Members'][i][2]
        n4 = df1['Members'][i][3]
        nlist1.append(n1)
        nlist2.append(n2)
        nlist3.append(n3)
        nlist4.append(n4)
    elif len(list) == 3:
        n1 = df1['Members'][i][0]
        n2 = df1['Members'][i][1]
        n3 = df1['Members'][i][2]
        n4 = 'Noname'
        nlist1.append(n1)
        nlist2.append(n2)
        nlist3.append(n3)
        nlist4.append(n4)
    elif len(list) == 2:
        n1 = df1['Members'][i][0]
        n2 = df1['Members'][i][1]
        n3 = 'Noname'
        n4 = 'Noname'
        nlist1.append(n1)
        nlist2.append(n2)
        nlist3.append(n3)
        nlist4.append(n4)
    elif len(list) == 1:
        n1 = df1['Members'][i][0]
        n2 = 'Noname'
        n3 = 'Noname'
        n4 = 'Noname'
        nlist1.append(n1)
        nlist2.append(n2)
        nlist3.append(n3)
        nlist4.append(n4)

df2['Group'] = group
df2['Member1'] = nlist1
df2['Member2'] = nlist2
df2['Member3'] = nlist3
df2['Member4'] = nlist4

df2
```
